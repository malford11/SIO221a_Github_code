**Transfer function:**

We discussed the fact that coherence is analogous to a correlation coefficient.
It tells us if two things vary in tandem in a consistent way, but it doesn't
tell us how big they are or how to use one variable to approximate a second
variable.  If we want to look at relative sizes, or if we want to approximate
a variable $y$ based on its relationship with $z$, in the time domain we think
about finding a regression coefficient.
We're used to solving a least-squares fitting equation of the form
\begin{equation}
{\bf y}=\bf{A}{\bf x}
\end{equation}
with a solution of the form
\begin{equation}
{\bf x} = {(\bf A}^T{\bf A})^{-1} {\bf A}^T {\bf y}
\end{equation}
For illustration we can simplify this to a case where the matrix ${\bf A}$ has
only column, that is where we regress ${\bf y'}$ (with the prime telling us that
the mean has been removed, since we don't want to complicate our least-squares fit)
against just one variable, making
${\bf A}$ a column vector (e.g. $\bf z'$).  In this case $x$ becomes a scalar and the matrix
inverse $({\bf A}^T{\bf A})^{-1}$ is just the reciprocal of the variance of $\bf z'$.
\begin{equation}
x = \frac{\langle y'z'\rangle}{\langle z'^2\rangle}.
\end{equation}
Compare this with the correlation coefficient of the demeaned variables
\begin{equation}
r =  \frac{\langle y'z'\rangle}{\sqrt{\langle y'^2\rangle \langle z'^2\rangle}}.
\end{equation}
Viewed in this way, the correlation coefficient $r$ and the regression coefficient $x$
(another term for the least-squares fit with only one variable) look nearly
the same, aside from the normalization.
Both the correlation coefficient and the regression coefficient convey
useful information.  And that might make you think that
in the Fourier-transform domain
there should be a form analogous to regression.

The term for the Fourier domain analog to regression is the transfer function:
\begin{equation}
\hat{H}_{zy}(f) = \frac{\hat{G}_{zy}(f)}{\hat{G}_{zz}(f)},
\end{equation}
which provides a (complex-numbered) recipe for mapping from $z$ to $y$.


Formally, we talk about the transfer function when we think about
constructing a linear system:
\begin{equation}
\mathcal{L}(y(t)) = z(t)
\end{equation}
If $\mathcal{L}$ is a linear operator, then we could think of this relationship
as a convolution:
\begin{equation}
y_t = \int_{-\infty}^{\infty} h(u) z(t-u)\, du
\end{equation}
or if we Fourier transform, this would state:
\begin{equation}
Y(f) = H(f) Z(f).
\end{equation}

This is the final analogy between time domain and frequency domain, so let's take a moment to summarize all of the tools we've covered up to now.

**Transfer functions (or gain functions):  a proper example**

If we want to look at relative sizes of two Fourier-transformed quantities, 
we can look at the
transfer function (also known as the gain function):
$$
\hat{H}_{zy}(f) = \frac{\hat{G}_{zy}(f)}{\hat{G}_{zz}(f)},
$$

which provides a (complex-numbered) recipe for mapping from $z$ to $y$.
Then, if 
\begin{equation}
y_t = \int_{-\infty}^{\infty} h(u) z(t-u)\, du,
\end{equation}
the Fourier transform is:
\begin{equation}
Y(f) = H(f) Z(f).
\end{equation}

Consider it this way.  Suppose we have a linear system like we discussed earlier in the quarter - it could be a damped pendulum for example: 
\begin{equation}
z(t) = \frac{d^2y}{dt^2} + \alpha \frac{dy}{dt} + \beta y
\end{equation}
Then by Fourier transforming, we have:
\begin{eqnarray}
Z(f) & = & -f^2 Y(f)+i\alpha f Y(f) + \beta Y(f) \\
 & = &  Y(f)\left[\beta -f^2 +i\alpha f\right]
\end{eqnarray}
so
\begin{equation}
Y(f) = \frac{1}{\left[\beta -f^2 +i\alpha f\right]} Z(f)
\end{equation}
and
\begin{equation}
H(f) = \frac{1}{\left[\beta -f^2 +i\alpha f\right]}
\end{equation}

This is a nice framework for solving differential equations, but can we
use it to gain insights into our data as well?


There are some nice rules for transfer functions:

1. Linearity:  If a given linear system has an input $z_1(t)$ which corresponds to an output $y_1(t)$, and input $z_2(t)$ corresponds to output $y_2(t)$m then a summed input $z(t)=\alpha z_1(t)+\beta z_2(t)$, will produce an output $y(t)=\alpha y_1(t)+\beta y_2(t)$.
2. Time invariance:  If an input is delayed in time by $\tau$, then the output is as well:  If $z(t) \rightarrow z(t+\tau)$, then $y(t) \rightarrow y(t+\tau)$.


3. Causality:  If h(t) represents an impulse, then it should be zero for $t<0$.  A response cannot occur before the forcing.

4. Sequential application:  If the output of one linear system is an input to a second system, then the frequency response is 
$$
H_{12}(f) = H_1(f)\cdot H_2(f)
$$

So suppose we measure $y(t)$ and $z(t)$.  Can we determine $h$ or $H$?
We know that
\begin{equation}
Y(f) = H(f)Z(f)
\end{equation}
Let's multiply both sides of the equation by the complex conjugate of $Z$ to
form the cross-spectrum:
\begin{equation}
\frac{Y(f)Z^*(f)}{\Delta f} = H(f)\frac{Z(f)Z^*(f)}{\Delta f}
\end{equation}
This becomes
\begin{equation}
G_{zy}(f)= H(f)G_{zz}(f)
\end{equation}
so
\begin{equation}
 H(f)=\frac{G_{zy}(f)}{G_{zz}(f)}
\end{equation}

**Instrument response example**

Say we have a variable $x(t)$ and an instrument that averages a little before recording the output: $\hat{x}(t)$.  This would then amount to a convolution of the instrument's response $h(\tau)$.  For example, a simple instrument might compute a running mean or, in the case of ADCPs, might use a triangle window in depth profiles.  In either case, we know the Fourier transform of the true and measured signals are simply related: $\hat{X}(f)=X(f)R(f)$ where $R(f)$ is the Fourier transform of the filter.  We would then say that the transfer function associated with the instrument response is 
$$
H(f)=\frac{\hat{X}(f)}{X(f)}=R^2(f).
$$

Note this is related to the coherence, but different in that it directly maps the true quantity onto the measured quantity.  In this case, it has no units since the measured and true quantities have the same units.

**Transfer function: Practical implication**

You might be wondering exactly how to compute a transfer function, so here's a practical example.

```matlab
lambda=10; % 10 m wavelength
V=0.3; % 0.3 m/s propagation
n2s=0.2; % noise-to-signal ratio
time=(1:5000)';
x=n2s*randn(5000,1)+cos(2*pi/lambda*V*time);
y=n2s*randn(5000,1)+cos(2*pi/lambda*V*(time)+pi/2);


% start with a segmented data approach:
xx=[reshape(x,500,10) reshape(x(251:4750),500,9)];
yy=[reshape(y,500,10) reshape(y(251:4750),500,9)];
fxx=fft(xx.*(hanning(500)*ones(1,19)));
fyy=fft(yy.*(hanning(500)*ones(1,19)));
sxx=abs(fxx(1:251,:)).^2;
syy=abs(fyy(1:251,:)).^2;
sxy=conj(fxx(1:251,:)).*fyy(1:251,:);
coher1=abs(mean(sxy,2)).^2./(mean(sxx,2).*mean(syy,2));
phase1=atan2(-imag(mean(sxy,2)),real(mean(sxy,2)));
subplot(211)
semilogx(0:250,coher1)

transfer=mean(sxy,2) ./mean(sxx,2);
semilogx(0:250,coher1,0:250,abs(transfer))
set(gca,'FontSize',14)
ylabel('coherence or transfer function','FontSize',14)
legend('coherence','transfer function amplitude','fontsize',8)
subplot(212)
semilogx(0:250,phase1)
grid
ylabel('phase','FontSize',14)

xlabel('frequency','FontSize',14)

```

**Salinity spiking example**

A classic example comes from the "salinity spiking" problem, which arises
due to the different response times of temperature and conductivity sensors:
\begin{equation}
S=S(T(t),C(t-\tau)).
\end{equation}
Our challenge now is to use the transfer (or gain) function to assess the
salinity spiking, and maybe even to correct it.

Conductivity is strongly dependent on temperature, and we have
to remove the temperature effect to determine salinity.  But the faster response times of conductivity sensors relative to temperature sensors are a source of confusion.


A typical correction might allow for errors both in the response time and a
direct time lag:
\begin{equation}
\tau \frac{d\hat{T}(t)}{dt} + \hat{T}(t) = T(t-L)
\end{equation}
(from Giles and McDougall, Deep-Sea Research, 1986)
and this suggests corrections both in the frequency and time domain, either by
minimizing phase differences or by maximizing correlation.
We can Fourier transform this to find:
\begin{equation}
-i2\pi f\tau {\cal{F}}(\hat{T}) + {\cal{F}}(\hat{T}) = {\cal{F}}(T)e^{-i2\pi fL},
\end{equation}
implying that
\begin{equation}
(1-i2\pi f\tau)e^{i2\pi f L} {\cal{F}}(\hat{T}) = {\cal{F}}(T).
\end{equation}
Since conductivity has the fast response, one strategy is to treat $\hat{C}$
as behaving like the true temperature $T$, with a scaling factor $\alpha$.  So hypothetically:
\begin{eqnarray}
 \hat{T}(k) & = & H_T(k) T(k) = \frac{1}{(1-i2\pi f\tau)e^{i2\pi f L}} T(k)\\
 \hat{C}(k) & = & H_C(k) T(k) = T(k)\\
\end{eqnarray}
implying a correction P(k) of the form:
\begin{equation}
P(k) = \frac{\alpha H_T(k)}{H_C(k)} = \alpha \frac{1}{(1-i2\pi f\tau)e^{i2\pi f L}}.
\end{equation}

Let's have a look in MATLAB to gain some intuition on the transfer function predicted by this model.

```matlab
%We'll do this in the frequency domain:
tau=0.1; %timescale of the instrument response in seconds
f=0:0.1:12; %up to 24 Hz
L=1/24; %time lag in sec

H=exp(-i*2*pi*f*L)./(1 - i*2*pi*f*tau);

subplot(211)
plot(f,abs(H))
ylabel('Measured / true amplitude (unitless)')
xlabel('Frequency / Hz')
grid
subplot(212)
plot(f,atan2(imag(H),real(H)))
ylabel('Phase / radians')
xlabel('Frequency / Hz')
grid
```

To compute the parameters $\tau$ and $L$, we would compute cross-spectra of our temperature and conductivity profiles.  Here's a basic procedure.

1. Identify a segment of the water column in which temperature and conductivity should be well behaved, with fluctuations due to temperature only.  (This isn't essential, but it will give us a good shot at unraveling the sensor response time issues that lead to salinity or density spiking.) We'll identify the true values as $T$ and $C$, and the measured values as $\hat{T}$ and $\hat{C}$.

2. Collect a lot of profiles of data. 
3. Now treat this as a linear system:
\begin{eqnarray}
 \hat{T}(k) & = & H_T(k) T(k)\\
 \hat{C}(k) & = & H_C(k) T(k)\\
\end{eqnarray}
Here $H_T(k)$ is the spatial/frequency response of the temperature sensor,
and $H_C(k)$ is the spatial/frequency response of the conductivity sensor. Note they are related via the profiling speed of the instrument $w$: $f=wk$.
The use of $T(k)$ in the conductivity equation might seem a little crazy,
but it's really important, since we're asserting that salinity is unimportant
in our (hypothetical) study region.

4. Compute spectra and cross spectra:
\begin{eqnarray}
 \hat{G}_{TT}(k) & = & \frac{\langle \hat{T}^*(k)\hat{T}(k)\rangle}{\Delta k}\\
 \hat{G}_{CC}(k) & = & \frac{\langle \hat{C}^*(k)\hat{C}(k)\rangle}{\Delta k}\\
 \hat{G}_{TC}(k) & = & \frac{\langle \hat{T}^*(k)\hat{C}(k)\rangle}{\Delta k}\\
\end{eqnarray} 
Then if we substitute in the expessions linking the observed values to the
true values we obtain:
\begin{eqnarray}
\frac{\langle\hat{T}^*(k)\hat{C}(k)\rangle}{\Delta k} & = & \frac{\langle(H_T(k)T(k))^* H_C(k)T(k)\rangle}{\Delta k} \\
 & = & [H_T^*(k)H_C(k)] \frac{\langle T^*(k) T(k)\rangle}{\Delta k} \\
\hat{G}_{TC}(k) & = &  [H_T^*(k)H_C(k)] G_{TT}(k)
\end{eqnarray}
The same applies for the temperature spectrum:
\begin{eqnarray}
 \hat{G}_{TT}(k) & = & \frac{\langle \hat{T}^*(k)\hat{T}(k)\rangle}{\Delta k}\\
 & = & [H_T^*(k)H_T(k)] G_{TT}(k)
\end{eqnarray}
So the ratio of these becomes:
\begin{equation}
\frac{\hat{G}_{TC}(k)}{\hat{G}_{TT}(k)} = \frac{H_T^*(k)H_C(k)}{|H_T|^2} =
\frac{H_C(k)}{H_T(k)}
\end{equation}
This means that even without knowing the response function $H$,
we can compute the ratio of the response functions
from the transfer function, the ratio of the  cross-spectrum to the
spectrum.  An analogous relationship also holds:
\begin{equation}
\frac{\hat{G}_{TC}(k)}{\hat{G}_{CC}(k)} = \frac{H_T^*(k)H_C(k)}{|H_C|^2} =
\frac{H_T(k)}{H_C(k)}
\end{equation}


5. Now use this information to correct the conductivity sensor to have the same response as the temperature sensor.  Here we'll define our corrected conductivity as $\hat{\hat{C}}(k)$, and we want to understand its relationship with the observed temperature $\hat{T}(k)$ and the true temperature $T(k)$.
\begin{equation}
\hat{\hat{C}}(k) = \alpha\hat{T}(k) = \alpha H_T(k) T(k)
\end{equation}
This means we need a correction of the form:
\begin{equation}
\hat{\hat{C}}(k) = \hat{C}(k)\cdot P(k) = \alpha\hat{T}(k),
\end{equation}
and our task is to figure out $P(k)$.
We can also write:
\begin{equation}
\hat{C}(k)\cdot P(k) = H_C(k)T(k)P(k)
\end{equation}
so putting this together:
\begin{equation}
\alpha H_T(k) T(k)  = H_C(k)T(k)P(k)
\end{equation}
Thus
\begin{equation}
P(k) = \frac{\alpha H_T(k)}{H_C(k)} = \alpha \frac{\hat{G}_{TC}(k)}{\hat{G}_{CC}(k)}
\end{equation}
where $\alpha$ is real.
So we do a bit of curve fitting to optimize our correction.

In this case, you'll see that the transfer function peak happens to be about the same size as the coherence, while the background noise for the transfer function is significantly larger than the background noise for the coherence.   These results are entirely a function of the data that we use.  The high background noise level stems from the fact that the coherence is scaled by $x$, which can be very small.
