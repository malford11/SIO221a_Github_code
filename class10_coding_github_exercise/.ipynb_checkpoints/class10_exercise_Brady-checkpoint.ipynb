```python
from scipy.io import loadmat 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import glob
import pandas as pd
from datetime import datetime
from datetime import timedelta
from matplotlib.dates import HourLocator 
from matplotlib.dates import DateFormatter
```

```python
file = loadmat('class10_record.mat') # loat .mat data file 
print(file)
```

```python
# convert the matlab data file to a python list 
data_list = [[element for element in upperElement] for upperElement in file['class10_record']]
print(data_list)
```

```python
# index through the list to extract the measurements and the time vector 
data = data_list[0][0][0][0]
time = data_list[0][0][1][0]
```

```python
# print the variables 
print(time)
print(data)
```

```python
# plot the data 
plt.figure(figsize=(8,6))
plt.plot(time,data)
plt.xlabel('Time',fontsize=16)
plt.ylabel('Mooring velocity $u$ [m/s]',fontsize=16)
plt.show()
```
